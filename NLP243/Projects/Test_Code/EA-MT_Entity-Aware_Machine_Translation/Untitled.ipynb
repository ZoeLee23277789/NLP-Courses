```python
import pandas as pd
import matplotlib.pyplot as plt

#  CSV 
file_path = "wandb_export_2024-12-01T01_12_58.117-08_00.csv"  # 
data = pd.read_csv(file_path)

# 
loss_max = data["/content/drive/MyDrive/mbart_finetuned_dynamic - train/loss__MAX"]
global_step = data["train/global_step"]

# 
plt.figure(figsize=(10, 6))
plt.plot(global_step, loss_max, label="Loss MAX", marker='o', linestyle='-')
plt.title("Loss MAX vs Global Step")
plt.xlabel("Global Step")
plt.ylabel("Loss MAX")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# 
file_path = './epoch_loss_results.csv'  # 
df = pd.read_csv(file_path)

# Train Loss  0.13  0.18  500 
filtered_df_train_loss = df[(df['Train Loss'] >= 0.13) & (df['Train Loss'] <= 0.18) & (df['Step'] % 500 == 0)]

# 
filtered_results_path = './filtered_train_loss_results.csv'  # 
filtered_df_train_loss.to_csv(filtered_results_path, index=False)

print(f" {filtered_results_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#  CSV 
file_path = "mBART_Loss.csv"  # 
data = pd.read_excel(file_path)

# 
data.columns = [col.strip() for col in data.columns]

# 
plt.figure(figsize=(10, 6))
plt.plot(data['Epoch'], data['Train Loss'], label='Train Loss', marker='o', linestyle='-')
plt.plot(data['Epoch'], data['Validation Loss'], label='Validation Loss', marker='o', linestyle='--')
plt.title("Train and Validation Loss vs Epoch")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.xticks(data['Epoch']) 
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
