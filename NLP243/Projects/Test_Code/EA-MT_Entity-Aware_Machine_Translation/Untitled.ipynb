```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取 CSV 文件
file_path = "wandb_export_2024-12-01T01_12_58.117-08_00.csv"  # 替換成你的文件路徑
data = pd.read_csv(file_path)

# 提取需要的數據
loss_max = data["/content/drive/MyDrive/mbart_finetuned_dynamic - train/loss__MAX"]
global_step = data["train/global_step"]

# 繪製圖表
plt.figure(figsize=(10, 6))
plt.plot(global_step, loss_max, label="Loss MAX", marker='o', linestyle='-')
plt.title("Loss MAX vs Global Step")
plt.xlabel("Global Step")
plt.ylabel("Loss MAX")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# 加載數據文件
file_path = './epoch_loss_results.csv'  # 請更新為您的文件路徑
df = pd.read_csv(file_path)

# 篩選條件：Train Loss 在 0.13 到 0.18 之間，並且步數為每 500 步
filtered_df_train_loss = df[(df['Train Loss'] >= 0.13) & (df['Train Loss'] <= 0.18) & (df['Step'] % 500 == 0)]

# 保存篩選結果到新文件
filtered_results_path = './filtered_train_loss_results.csv'  # 輸出文件路徑
filtered_df_train_loss.to_csv(filtered_results_path, index=False)

print(f"篩選後的結果已保存到 {filtered_results_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取 CSV 文件
file_path = "mBART_Loss.csv"  # 替換為您的文件路徑
data = pd.read_excel(file_path)

# 確保列名一致，去掉多餘空格
data.columns = [col.strip() for col in data.columns]

# 繪製圖表
plt.figure(figsize=(10, 6))
plt.plot(data['Epoch'], data['Train Loss'], label='Train Loss', marker='o', linestyle='-')
plt.plot(data['Epoch'], data['Validation Loss'], label='Validation Loss', marker='o', linestyle='--')
plt.title("Train and Validation Loss vs Epoch")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.xticks(data['Epoch']) 
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
