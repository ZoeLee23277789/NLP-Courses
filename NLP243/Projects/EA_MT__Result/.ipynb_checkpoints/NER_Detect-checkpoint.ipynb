```python
import hanlp

# 加載實體識別模型
recognizer = hanlp.load(hanlp.pretrained.ner.MSRA_NER_BERT_BASE_ZH)
print("HanLP 模型加載成功！")
```

```python
# 示例文本
text = "几年 前 ， 在 TED 大会 上 ， Peter Skillman 介绍 了 一个 设计 挑战"

# 識別實體
entities = recognizer(text)
print(entities)
```

```python
import json

# 請確保文件已上傳到 Colab
file_path = './translated_results_iwslt_1130.json'

# 加載 JSON 文件
with open(file_path, 'r', encoding='utf-8') as file:
    data = json.load(file)
```

```python
# 為每個句子識別 Reference Text 和 Best Translation 的實體
for entry in data:
    reference_text = entry["Reference Text"]
    best_translation = entry["Best Translation"]
    
    # 識別實體
    entry["Reference Entities"] = recognizer(reference_text)
    entry["Best Translation Entities"] = recognizer(best_translation)
```

```python
# 初始化統計變數
total_sentences = len(data)
matched_count = 0

# 計算匹配比例
for entry in data:
    ref_entities = set(tuple(entity) for entity in entry["Reference Entities"])
    best_entities = set(tuple(entity) for entity in entry["Best Translation Entities"])
    
    # 檢查 Reference 和 Best Translation 是否完全匹配
    if ref_entities == best_entities:
        matched_count += 1

# 計算比例
matched_ratio = matched_count / total_sentences
print(f"總句子數: {total_sentences}")
print(f"完全匹配的句子數: {matched_count}")
print(f"匹配比例: {matched_ratio:.2%}")
```

```python
# 保存實體識別結果到文件
output_path = './entity_recognition_results.json'

with open(output_path, 'w', encoding='utf-8') as outfile:
    json.dump(data, outfile, ensure_ascii=False, indent=4)

print(f"結果已保存到 {output_path}")
```
